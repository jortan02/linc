```python
import pandas as pd

# Drop rows with errors
evaluation_df = pd.read_json("FOLIO/data/v0.0/folio-validation.jsonl", lines=True)
error_indices = [
    3,
    109,
    110,
    111,
    6,
    28,
    30,
    48,
    113,
    115,
    139,
    140,
    10,
    11,
    12,
    88,
    106,
    107,
    108,
    174,
    175,
    176,
]
evaluation_df = evaluation_df.drop([x - 1 for x in error_indices])

# Convert lists to strings
evaluation_df["premises"] = evaluation_df["premises"].apply(lambda x: "\n".join(x))
evaluation_df["premises-FOL"] = evaluation_df["premises-FOL"].apply(lambda x: "\n".join(x))

evaluation_df = evaluation_df.reset_index(drop=True)
```

```python
evaluation_df.to_csv("data/folio.csv", index=False)
```
