```python
# Add new lines to theory premises
evaluation_df["premises"] = evaluation_df["theory"].apply(lambda row: ".\n".join(row.split(". ")))

# Expand the questions
evaluation_df = evaluation_df.set_index(["id", "premises"])["questions"].apply(pd.Series).stack().reset_index(name="question_info")
evaluation_df = evaluation_df.rename(columns={"level_2": "question_id"})

# Retrieve question info
evaluation_df["conclusion"] = evaluation_df["question_info"].apply(lambda row: row["question"])
evaluation_df["label"] = evaluation_df["question_info"].apply(lambda row: row["answer"])
evaluation_df["label"] = evaluation_df["label"].replace("Unknown", "Uncertain")
evaluation_df = evaluation_df.drop("question_info", axis=1)

# Sample the dataframe:
evaluation_df = evaluation_df.sample(n=182, replace=False, random_state=0)

evaluation_df = evaluation_df.reset_index(drop=True)
```
